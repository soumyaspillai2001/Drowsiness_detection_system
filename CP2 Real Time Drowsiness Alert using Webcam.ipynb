# Live Drowsiness Alert System with Alarm

## Initialize pygame for sound alert

```python
# Initialize pygame for sound alert
pygame.mixer.init()
alert_sound = pygame.mixer.Sound('alarm.wav')
```

## Load the trained CNN model

```python
# Load the trained model
model = load_model('drowsiness_detection_model.h5')
```

## Load the pre-trained models provided by open CV for image recognitions tasks like face,eye detection
- Note download these haarcascade xml files from open CV github repository

```python
face = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
leye = cv2.CascadeClassifier('haarcascade_lefteye_2splits.xml')
reye = cv2.CascadeClassifier('haarcascade_righteye_2splits.xml')
```

```python
#model1 = load_model('cnnDrowsinessDetection.h5')
```

```python
#lbl=['Close','Open']

#model = load_model('models/cnncat2.h5')
#path = os.getcwd()
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
count=0
score=0
thicc=2
rpred=[99]
lpred=[99]

while(True):
    ret, frame = cap.read()
    height,width = frame.shape[:2] 

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = face.detectMultiScale(gray,minNeighbors=5,scaleFactor=1.1,minSize=(25,25))
    left_eye = leye.detectMultiScale(gray)
    right_eye =  reye.detectMultiScale(gray)

    cv2.rectangle(frame, (0,height-50) , (200,height) , (0,0,0) , thickness=cv2.FILLED )

    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y) , (x+w,y+h) , (100,100,100) , 1 )

    for (x,y,w,h) in right_eye:
        r_eye=frame[y:y+h,x:x+w]
        count=count+1
        r_eye = cv2.cvtColor(r_eye,cv2.COLOR_BGR2GRAY)
        r_eye = cv2.resize(r_eye,(80,80))
        r_eye= r_eye/255
        r_eye=  r_eye.reshape(80,80,-1)
        r_eye = np.expand_dims(r_eye,axis=0)
        rpred = model.predict(r_eye)
        rpred = np.round(rpred).astype(int)
        if(rpred[0]==0):
            lbl='Open' 
        if(rpred[0]==1):
            lbl='Closed'
        break

    for (x,y,w,h) in left_eye:
        l_eye=frame[y:y+h,x:x+w]
        count=count+1
        l_eye = cv2.cvtColor(l_eye,cv2.COLOR_BGR2GRAY)  
        l_eye = cv2.resize(l_eye,(80,80))
        l_eye= l_eye/255
        l_eye=l_eye.reshape(80,80,-1)
        l_eye = np.expand_dims(l_eye,axis=0)
        lpred = model.predict(l_eye)
        lpred = np.round(lpred).astype(int)
        if(lpred[0]==1):
            lbl='Closed'   
        if(lpred[0]==0):
            lbl='Open'
        break

    if(rpred[0]==1 and lpred[0]==1):
        score=score+1
        cv2.putText(frame,"Closed",(10,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    # if(rpred[0]==1 or lpred[0]==1):
    else:
        score=score-1
        cv2.putText(frame,"Open",(10,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    
        
    if(score<0):
        score=0   
    cv2.putText(frame,'Score:'+str(score),(100,height-20), font, 1,(255,255,255),1,cv2.LINE_AA)
    if(score>5):
        #person is feeling sleepy so we beep the alarm
        #cv2.imwrite(os.path.join(path,'image.jpg'),frame)
        try:
            alert_sound.play()
            
        except:  # isplaying = False
            pass
        if(thicc<16):
            thicc= thicc+2
        else:
            thicc=thicc-2
            if(thicc<2):
                thicc=2
        cv2.rectangle(frame,(0,0),(width,height),(0,0,255),thicc) 
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Working of Live Drowsiness Alert System:
- The webcam video capture opens and each frame is captured and saved.
- The BGR image is converted into grayscale i.e. 1D as haarcascade classifier works on grayscale images.
- First the face classifier is going to detect the face and get the coordinates and draw a rectangle for the face.
- Then the left eye and right eye classifier is detecting individual eyes and return those coordinates
- The image of left and right eyes are converted into grayscale and preprocessed before using model for prediction.
- The eye image is passed to the model to predict eye status i.e.Close for Class 1 and Open for Class 0. And labels are assigned.
- The score is 0 when the eyes are open and the score increases when both the eyes are closed for few seconds.
- After a threshold reaches the alert sound is played and the sound stops when the score is 0 and eyes are open.
